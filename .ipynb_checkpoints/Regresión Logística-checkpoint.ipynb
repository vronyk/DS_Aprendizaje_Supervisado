# Aplicando el algoritmo de Regresión  Logística para resolver un problema de clasificación utilizando el lenguaje Python.

```python
#carga de librerias
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Cargue los datos y revision de formatos de cada variable.
Codificación de las variables como numéricas o factores según corresponda.

```python
# Load the rrhh csv
df = pd.read_csv('rrhh.csv')
# Drop the rows with missing values
df.dropna(axis=1,inplace=True)
# encoding variables 
df['Sexo'] = df['Sexo'].map({'Male':0,'Female':1})
df['Edad'] = df['Edad'].astype(int)
df['Salario'] = df['Salario'].astype(int)
df['Departamento'] = df['Departamento'].map({
                                             'Admin Offices':1,
                                             'Executive Office':2,
                                             'IT/IS':3,
                                             'Production':4,
                                             'Sales':5,
                                             'Software Engineering':6
})

print(df.info()) #obtener detalle de las variables
print(df.describe(include='all')) #Descriptivos numéricos
df.head()
```

Realice un análisis descriptivo de sus datos.
Presencia de valores atípicos en las variables.

```python
df = df.drop(['Posicion','Desempeño'], axis=1)
df.head()
# Converitir las variables categoricas en variables dummy
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
plt.subplot(3,3,1)
plt.boxplot(df['Estado'] )
plt.title("Estado")

plt.subplot(3,3,2)
plt.boxplot(df['Edad'] )
plt.title("Edad")

plt.subplot(3,3,3)
plt.boxplot(df['Ratio.Pago'] )
plt.title("Ratio.Pago")
					
plt.subplot(3,3,4)
plt.boxplot(df['Dias.trabajados'] )
plt.title("Dias.trabajados")

plt.subplot(3,3,5)
plt.boxplot(df['Ausencias'] )
plt.title("Ausencias")

plt.subplot(3,3,6)
plt.boxplot(df['Sexo'] )
plt.title("Sexo")

plt.subplot(3,3,7)
plt.boxplot(df['Estado.Civil'] )
plt.title("Estado.Civil")

plt.subplot(3,3,8)
plt.boxplot(df['Departamento'])
plt.title("Departamento")

plt.subplot(3,3,9)
plt.boxplot(df['Salario'])
plt.title("Salario")

plt.tight_layout() #para que no se superpongan las gráficas
#Mostar los graficos
plt.show()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Estado', axis=1),
    df['Estado'],
    test_size=0.2, random_state=1234
)
# Create the logistic regression model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train, y_train)
# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Compute the confusion matrix
from sklearn.metrics import confusion_matrix
cf_matrix = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(4, 3))
sns.heatmap(
    cf_matrix, annot=True, cmap="Blues", fmt="d", ax=ax
)
ax.set_xlabel('Predicted')
ax.set_ylabel('True')
ax.set_title('Confusion Matrix')
```
